```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(n_samples=200, centers=3, cluster_std=2.7, random_state=42)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Synthetic Data with 3 Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, linewidths=3, color='r')
plt.title("K-means Clustering with K=3")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

lowest_sse = float('inf')
final_centroids = None
num_iterations_to_converge = None

for i in range(10):
  
    kmeans = KMeans(n_clusters=3, init='random', n_init=1, max_iter=300, random_state=i)
    
    
    kmeans.fit(X)
    
    
    sse = kmeans.inertia_
    
    
    if sse < lowest_sse:
        lowest_sse = sse
        final_centroids = kmeans.cluster_centers_
        num_iterations_to_converge = kmeans.n_iter_
    
    print(f"Variation {i+1} SSE: {sse:.2f}")

print(f"\nLowest SSE value: {lowest_sse:.2f}")
print(f"Final centroids: {final_centroids}")
print(f"Number of iterations to converge: {num_iterations_to_converge}")
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], s=50);
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()

predicted_labels = kmeans.predict(X[:10])
print("\nPredicted labels for first 10 points:")
print(predicted_labels)
```

```python
from sklearn.cluster import KMeans
import numpy as np
import warnings
warnings.filterwarnings('ignore')

wcss_values=[]

k_values = range(1, 10)


for k in k_values:
    model = KMeans(n_clusters=k)
    model.fit(X)
    wcss_values.append(model.inertia_)

plt.plot(k_values, wcss_values, 'o-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings('ignore')


k_values = range(2, 10)


silhouette_scores = []

for k in k_values:
    model = KMeans(n_clusters=k)
    labels = model.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)


plt.plot(k_values, silhouette_scores, 'o-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method')
plt.show()
```
