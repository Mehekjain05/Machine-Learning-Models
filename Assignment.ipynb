```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/datasets/mumbai 2023-01-01 to 2023-04-30.csv')
df1.head(6)
```

```python
df1.info()
```

```python
df1['datetime'] = pd.to_datetime(df1['datetime'])
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1.drop(['snowdepth'], axis=1, inplace=True)
```

```python
df1['preciptype'].unique()
```

```python
df1['preciptype']= df1['preciptype'].replace({np.nan:"no rain"})
df1['preciptype'].unique()
```

```python
sns.heatmap(df1.isnull(), yticklabels=False)
```

```python
df12 = pd.read_csv('/content/drive/MyDrive/datasets/AQI.csv')
df12.head(6)
```

```python
df12['PM2.5'] = df12['PM2.5'].replace({"None": np.nan})
df12.isnull().sum()
```

```python
df12= df12.astype({'PM2.5':'double'})

mean= df12['PM2.5'].mean()

df12['PM2.5'] = df12['PM2.5'].replace({np.nan:mean})

df12['PM2.5'].isnull().sum()
```

```python
df12['From Date'] = pd.to_datetime(df12['From Date'])
df12.rename(columns={'From Date':'datetime'},inplace=True)
df12.head()
```

```python
df = pd.merge(
  df1, df12, on=["datetime"]
)
df.head(2)
```

```python
from mpl_toolkits import mplot3d
sns.set(style='darkgrid')

fig = plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d')

x=df['temp']
y=df['solarradiation']
z=df['PM2.5']

ax.scatter3D(x, y, z, color='blue')
ax.set_title("temp vs Solar radiation vs PM2.5", pad=25, size=15)
ax.set_xlabel('temp') 
ax.set_ylabel('Solar radiation') 
ax.set_zlabel('PM2.5')
```

```python
fig = plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d')

x=df['solarradiation']
y=df['dew']
z=df['SO2']

ax.scatter3D(x, y, z, color='blue')
ax.set_title("Solar radiation vs dew vs SO2", pad=25, size=15)
ax.set_xlabel('solar radiation') 
ax.set_ylabel('dew') 
ax.set_zlabel('SO2')
```

```python
sns.heatmap((df.corr())*100)
```

```python
x=df[['solarradiation','dew','tempmin','cloudcover','moonphase','humidity']]
y=df[['PM2.5']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import mean_squared_error

X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

reg = LinearRegression()

reg.fit(X_train,y_train)

y_pred = reg.predict(X_test)

mse = mean_squared_error(y_test,y_pred)
mse
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import export_graphviz
import graphviz

X = df[['solarradiation','dew','temp','cloudcover','windspeed','moonphase','humidity']] 
Y = df['PM2.5']

rf = RandomForestRegressor(n_estimators=100, max_depth=8, min_samples_split=2, random_state=1)

rf.fit(X, Y)

dot_df = export_graphviz(rf.estimators_[0], feature_names=['temp','windspeed','cloudcover','sealevelpressure','solarradiation','uvindex','dew'],  
                           filled=True, rounded=True)  

graph = graphviz.Source(dot_df, format='png')
graph.render("PM2.5_tree") 
```

```python
from sklearn import tree
tree.plot_tree(rf.estimators_[0])
```

```python
rf.score(X,Y)
```
